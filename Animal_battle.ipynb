```python
import random

# define animal class
class Animal:
    def __init__(self, name, power):
        self.name = name
        self.power = power

# def player class        
class Player:
    def __init__(self, name):
        self.name = name
        self.animals = []
        self.score = 0

    def add_animal(self, animal):
        self.animals.append(animal)

    def play_animal(self):
        return self.animals.pop(0)

# game data
animal_names = ["tiger", "lion", "Elephant", "Cheetah", "Wolf", "Eagle", "Crocodile", "Panther"]
animal_power = [90, 85, 95, 80, 70, 65, 88, 77]

# create and shuffle animals
all_animals = [Animal(name, power) for name, power in zip(animal_names, animal_power)]
random.shuffle(all_animals)

# split animals between players
player_one_animals = all_animals[:4]
player_two_animals = all_animals[4:]

# create player objects
player1 = Player("Player 1")
player2 = Player("Player 2")

for animal in player_one_animals:
    player1.add_animal(animal)

for animal in player_two_animals:
    player2.add_animal(animal)

print("⚔️ Animal Battle Begins!\n")

# game loop
for round_num in range(4):
    print(f"Round {round_num + 1} :")
    a1 = player1.play_animal()
    a2 = player2.play_animal()
    
    print(f"{player1.name} plays {a1.name} with power {a1.power}")
    print(f"{player2.name} plays {a2.name} with power {a2.power}")
    
    if a1.power > a2.power:
        print(f"{player1.name} wins this round\n")
        player1.score += 1
    elif a2.power > a1.power:
        print(f"{player2.name} wins this round\n")
        player2.score += 1
    else:
        print("It's a tie this round\n")

# final results
print(" Final Score:")
print(f"{player1.name}: {player1.score}")
print(f"{player2.name}: {player2.score}")

if player1.score > player2.score:
    print(f"\n {player1.name} wins the game!")
elif player2.score > player1.score:
    print(f"\n {player2.name} wins the game!")
else:
    print("\n It's a tie!")
```
